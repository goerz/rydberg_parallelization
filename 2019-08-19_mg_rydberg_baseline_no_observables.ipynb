# Baseline Rydberg Hamiltonian (no observables)

This notebook establishes a benchmarking benchmark by propagating a large Rydberg Hamiltonian with the current `master` version of QDYN (single-thread). Compared to [`2019-08-19_mg_rydberg_baseline.ipynb`](2019-08-19_mg_rydberg_baseline.ipynb), it omits calculating and writing out any expectation values. Thus, we can determine how much overhead the calculation of expectation values causes.

```python
import qutip
import numpy as np
import matplotlib.pylab as plt
```

```python
from src.qdyn.io import read_indexed_matrix
from src.qdyn.model import LevelModel
from src.qdyn.pulse import Pulse
from src.plots import show_spy
```

```python
%matplotlib inline
```

```python
V_per_cm = 1.0 / 5.14220624463189208e09
```

```python
# 1-based indices of the initial and target states
N_INITIAL = 1423
N_TARGET = 2696
# size of Hilbert space
N_HILBERT = N_TARGET
```

```python
def rydberg_hamiltonian(
    H_drift_file,
    H_sigma_file,
    H_pi_file,
    F_DC,
    Omega_sigma,
):
    """Construct the time-dependent Hamiltonian

    Args:
        H_drift_file (str)
        H_sigma_file (str)
        H_pi_file (str)
        F_DC (float)
        Omega_sig (Pulse)
        exapnd_H_sigma (bool)
    """
    H_drift = qutip.Qobj(
        read_indexed_matrix(
            H_drift_file, shape=(N_HILBERT, N_HILBERT), expand_hermitian=False
        )
    )
    H_sigma = qutip.Qobj(
        read_indexed_matrix(
            H_sigma_file,
            shape=(N_HILBERT, N_HILBERT),
            expand_hermitian=False,
        )
    )
    H_pi = qutip.Qobj(
        read_indexed_matrix(
            H_pi_file, shape=(N_HILBERT, N_HILBERT), expand_hermitian=False
        )
    )
    H_0 = H_drift + F_DC * H_pi
    return [H_0, [H_sigma, Omega_sigma], [H_sigma.dag(), Omega_sigma]]
```

```python
H = rydberg_hamiltonian(
    'data/ham_drift.dat',
    'data/ham_sig.dat',
    'data/ham_pi.dat',
    F_DC=2.5 * V_per_cm,
    Omega_sigma=Pulse.read('./data/pulse_sig.dat')
)
```

```python
psi0 = qutip.ket([N_INITIAL-1], dim=H[0].shape[0])
psi_target = qutip.ket([N_TARGET-1], dim=H[0].shape[0])
```

```python
show_spy(H[0].data, show_diagonal=False)
```

```python
show_spy((H[1][0].data + H[2][0].data), show_diagonal=True)
```

```python
H[1][1].show()
```

```python
qdyn_model = LevelModel()
```

```python
qdyn_model.add_ham(H[0], op_unit='au', op_type='pot')
```

```python
qdyn_model.add_ham(H[1][0], pulse=H[1][1], op_unit='au', op_type='dip')
```

```python
qdyn_model.add_ham(H[2][0], pulse=H[1][1], op_unit='au', op_type='dip', conjg_pulse=True)
```

```python
qdyn_model.add_state(psi0, label='initial')
```

```python
qdyn_model.add_state(psi_target, label='target')
```

```python
qdyn_model.set_propagation(
    T=H[1][1].T,
    nt=len(H[1][1].states_tgrid),
    time_unit=H[1][1].time_unit,
)
```

```python
qdyn_model.write_to_runfolder('./rf_baseline_no_observables')
```
